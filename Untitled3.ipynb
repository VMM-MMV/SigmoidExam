Here we saw that Year_of_Release and Publisher have some NAN's, and I could remove them or replace them, but I am going to replace them with their actual value by using ChatGPT 4 and it's web search capabilities, to search the year and publisher for each NAN.

Now I made a new data frame with all of the games where the year of release is missing.

```python
missing_year_rows = df[df["Year_of_Release"].isna()]
missing_year_rows.to_csv('MissingYearOfRelease.csv', index=False)
missing_year_rows
```

Now we do the same for the Publisher column.

There are some problems with this column since some NaN's are represented with "Unknown" instead.

We can check this by

```python
df["Publisher"].isna().sum()
```

```python
for i in df.columns:
    try:
      print(i, df[i].value_counts()["Unknown"])
    except:
      pass
```

```python
df["Publisher"].replace("Unknown", np.nan, inplace=True)
for i in df.columns:
    try:
      print(i, df[i].value_counts()["Unknown"])
    except:
      pass

df["Publisher"].isna().sum()
```

```python
missing_rows = df[df["Year_of_Release"].isna() | df["Publisher"].isna()]

# Save the missing rows to a CSV file
missing_rows.to_csv('MissingRows.csv', index=False)

# Display the missing rows
print(missing_rows.len())
```
